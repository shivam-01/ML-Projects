# Logistic Regression

'''To Classify iris species within versicolor or not  (Binary Classification).'''

```python
# Importing Necessary Libraries
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Reading Data
data = pd.read_csv('iris.csv')
print(data.shape)
data.head()
```

```python
# Visualizing
sns.countplot(x = 'species', data = data)
```

```python
# Sigmoid Function
def sigmoid(z):
    return 1 / (1 + np.exp(-z))
```

```python
# Hypothesis
def predict(X, weights):
    z = np.dot(X, weights)
    return sigmoid(z)
```

```python
# Cost Function
def cost_function(X, Y, weights):
    hypothesis = predict(X, weights)
    cost = -np.mean(Y * np.log(hypothesis) + (1 - Y) * np.log(1 - hypothesis))
    return cost
```

```python
# Gradient Descent
def update_weights(weights, X, Y, alpha, epochs):
    m = len(Y)
    hypothesis = predict(X, weights)
    loss = hypothesis - Y
    gradient = X.T.dot(loss) / m
    weights = weights - alpha * gradient
    cost = cost_function(X, Y, weights)
    
    # Log Progress
    if epoch % 10 == 0:
        print("epoch : {:d}    cost = {}".format(epoch, cost))
    
    return weights
```

```python
# Training
x1 = data['sepal_length'].values
x2 = data['sepal_width'].values
x0 = np.ones(len(x1))

X = np.array([x0, x1, x2]).T

data['Species'] = data['species'].apply(lambda x : 1 if x == "setosa" else 0)
Y = data['Species'].values

weights = np.array([1, 1, 1])

alpha = 0.01
epochs = 10000
prob = []
for epoch in range(epochs) :
    weights = update_weights(weights, X, Y, alpha, epochs)
    prob = predict(X, weights)
print(weights)
```

```python
# Target Class
preds = []
for i in prob:
    if i >= 0.5:
        preds.append(1)
    else:
        preds.append(0)
```

```python
# Accuracy
from sklearn.metrics import f1_score
f1_score(preds, Y)
```

```python
# Visualizing
plt.scatter(x1, x2, c = Y.ravel())
ax = plt.gca()
xvals = np.array(ax.get_xlim()).reshape(-1,1)
yvals = -(xvals * weights[1] + weights[0]) / weights[2]
plt.plot(xvals, yvals)
```

```python
# Train and Test Data
x = data.drop('species', axis = 1)
y = Y

from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.1, random_state = 1)
```

```python
# Logistic Regression
from sklearn.linear_model import LogisticRegression
logmodel = LogisticRegression(solver='liblinear')
logmodel.fit(x_train,y_train)
predictions = logmodel.predict(x_test)
```

```python
# Classification
from sklearn.metrics import classification_report
print(classification_report(y_test, predictions))
```

```python
# Confusion Matrix
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test, predictions)
```

```python
# f1_score
from sklearn.metrics import f1_score
f1_score(y_test, predictions)
```
